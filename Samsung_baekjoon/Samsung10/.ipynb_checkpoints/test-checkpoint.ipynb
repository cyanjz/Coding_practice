```python
ns = [_ for _ in range(10)]
# 2 3 3 2
N = 10
temp = [2, 3, 3, 2]
dims = []
for _ in range(4):
    dim = 1
    for i in range(temp[_]):
        dim *= (N-i)
    for i in range(1, temp[_]+1):
        dim /= i
    N -= temp[_]
    dims.append(int(dim))
print(dims)
```
